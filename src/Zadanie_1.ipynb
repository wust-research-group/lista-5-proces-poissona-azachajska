```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generuj_czasy_oczekiwania(T, lambd):
    t = 0  
    czasy_skokow = []
    Nt = []

    while t < T:
        U = np.random.rand()
        t -= 1 / lambd * np.log(U)
        if t < T:
            czasy_skokow.append(t)
            Nt.append(len(czasy_skokow))
    return czasy_skokow, Nt
```

```python
def generuj_momenty_skokow(T, lambd):
    n = np.random.poisson(lambd * T)
    Nt = []
    if n == 0:
        return [], Nt
    skoki = np.random.uniform(0, T, n)
    skoki.sort()

    for i in range(1, n + 1):
        Nt.append(i)
    return skoki, Nt
```

```python
lam = 1  
T = 100  

times1, Nt1 = generuj_czasy_oczekiwania(T, lam)
times2, Nt2 = generuj_momenty_skokow(T, lam)

plt.step(times1, Nt1, label='Metoda 1')
plt.step(times2, Nt2, label='Metoda 2')
plt.xlabel('Czas')
plt.ylabel('Nt')
plt.title('Trajektorie procesu Poissona')
plt.legend()
plt.show()
```
