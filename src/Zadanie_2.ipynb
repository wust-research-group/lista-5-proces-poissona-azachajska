```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def kompensowany_poisson(lambda_, T, num_points):
    z = np.random.exponential(1/lambda_, num_points)
    z = np.cumsum(z)

    times = np.linspace(0, T, num_points)
    compens = z - lambda_ * np.arange(num_points) * (T / num_points)
    return times, compens
```

```python
lambda_ = 0.5 
T = 10  
num_points = 1000  

time, trajectory = kompensowany_poisson(lambda_, T, num_points)

plt.plot(time, trajectory)
plt.show()
```
